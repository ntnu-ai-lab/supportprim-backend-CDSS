```python
def debug_print(codebook_id='', infopad_group_txt='', infopad_question_txt='', infopad_answere_txt='', ism_group_var='', ism_question_var='', ism_answere_var='', ism_variable=''):
    print(
        codebook_id,          '   :  \t', 
        infopad_group_txt,    ' \t:  \t', 
        infopad_question_txt, ' \t:  \t', 
        infopad_answere_txt,  ' \t:  \t', 
        ism_group_var,        ' \t:  \t', 
        ism_question_var,     ' \t:  \t', 
        ism_answere_var,      ' \t:  \t', 
        ism_variable) 
```

```python
def condition_df_appender(
    df,
    codebook_id, 
    infopad_group_txt, 
    infopad_question_txt, 
    infopad_answer_txt, 
    ism_group_var, 
    ism_question_var, 
    ism_answer_var, 
    ism_variable):
    
    df = df.append(
                    {
                        'codebook_id'           : codebook_id,               
                        'infopad_group_txt'     : infopad_group_txt,              
                        'infopad_question_txt'  : infopad_question_txt,              
                        'infopad_answer_txt'    : infopad_answer_txt,                
                        'ism_group_var'         : ism_group_var,                
                        'ism_question_var'      : ism_question_var,               
                        'ism_answer_var'        : ism_answer_var,              
                        'ism_variable'          : ism_variable           
                   }, 
                   ignore_index=True)
    return df
```

```python
def gen_answer_xslt(df, empty_val='_unknown_'):
    for infopad_answer_txt, ism_answer_var in zip(df.infopad_answer_txt, df.ism_answer_var):
        #print ( infopad_answer_txt, ism_answer_var)

        print (
            "<xsl:if test=\"answer[@selected='true'] and answer='"+infopad_answer_txt+"'\">\n"+
            "    <"+ism_answer_var+"><xsl:value-of select=\"answer[@selected='true']\"/></"+ism_answer_var+">\n"+
            "</xsl:if>\n"
        )

gen_answer_xslt(condition_df_list[1][0:1])
```

```python
def gen_question_xslt(df, empty_val='_unknown_'):
    for infopad_question_txt, ism_question_var in zip(df.infopad_question_txt, df.ism_question_var):
        print (
            "<xsl:if test=\"@name='"+infopad_question_txt+"'\">\n"+
            "    <"+ism_question_var+">\n"+
            "        <xsl:choose>\n"+
            "            <xsl:when test=\"answer/@selected='true'\">\n"+
            "                <xsl:value-of select=\"answer[@selected='true']\"/>\n"+
            "            </xsl:when>\n"+
            "            <xsl:when test=\"answer[not(@selected)]\">\n"+
            "                <xsl:value-of select=\"answer\"/>\n"+
            "            </xsl:when>\n"            
            "            <xsl:otherwise>"+empty_val+"</xsl:otherwise>\n"+
            "        </xsl:choose>\n"+
            "    </"+ism_question_var+">\n"+
            "</xsl:if>\n"
        )
gen_question_xslt(condition_df_list[2][0:1])
```
