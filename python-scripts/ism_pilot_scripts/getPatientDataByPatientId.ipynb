# Get the complete patient data from Elasticsearch

```python
es_base_url = 'http://localhost:9200'
headers = {'Content-Type': 'application/json'}
```

```python
import pandas as pd
import numpy as np
import requests
import json

import os
os.getcwd()
```

```python
es_main_index = 'manual_patients'
es_trace_index = 'manual_patient_traces'
```

```python
def get_patient_by_pid( pid, es_index=es_main_index):
    data = '{  "query": {    "term": {      "patientId.keyword": "'+pid+'"    }  }}'
    response = requests.get(f'{es_base_url}/{es_index}/_search?filter_path=hits.hits._source', headers=headers, data=data)
    response_json = response.json().get('hits', {}).get('hits', [{}])[0].get('_source')
    return json.dumps(response_json, sort_keys=True, indent=2)

def patient_details_by_pid(pid):
    print(get_patient_by_pid(pid))
```
